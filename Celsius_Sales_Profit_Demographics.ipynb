```python
# Celsius_Data_Analysis.ipynb
# Step-by-Step Dataset Creation and Basic Analysis

import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Build the data tables ---

# 1A. Sales & Revenue Data (in USD millions)
sales_data = pd.DataFrame({
    "Period": ["FY 2024", "Q3 2025", "Trailing Retail Sales"],
    "Revenue": [1360, 725.1, 4000],  # FY Revenue & Q3 & rolling retail sales
    "Notes": [
        "Annual revenue for 2024 (approx.)",
        "Quarterly revenue Q3 2025",
        "Trailing 52-week retail sales"
    ]
})

# 1B. Profit & Margin Data
margin_data = pd.DataFrame({
    "Metric": ["Gross Margin %", "Operating Margin %"],
    "Value": [51.3, 4.33],
    "Description": [
        "Celsius gross margin (Q3 2025)",
        "Trailing operating margin (TTM) for Celsius"
    ]
})

# 1C. Consumer Demographics
demographics_data = pd.DataFrame({
    "Group": ["18-34 yrs", "35-55 yrs", "Male", "Female", "Regular gym/activity"],
    "Percentage (%)": [60, 30, 55, 45, 70],
    "Source": [
        "Core Celsius consumer age segment",
        "Secondary age segment",
        "Estimated gender split",
        "Estimated gender split",
        "Consumers reporting active lifestyles"
    ]
})

# 1D. Market Share (U.S. Energy Drinks)
market_share_data = pd.DataFrame({
    "Brand": ["Celsius (CELH)", "Alani Nu", "Monster", "Red Bull (overall)"],
    "Market Share (%)": [20.8, 6.3, 35, 35],
    "Notes": [
        "Celsius share in U.S. energy drink category (Q3 2025)",
        "Alani Nu share (2025)",
        "Estimate for competitor",
        "Estimate for competitor"
    ]
})

# Display the tables
print("=== Sales & Revenue ===")
display(sales_data)

print("=== Profit & Margin ===")
display(margin_data)

print("=== Consumer Demographics ===")
display(demographics_data)

print("=== Market Share Comparison ===")
display(market_share_data)

# --- 2. Visualization ---

# Plot Revenue Table
plt.figure(figsize=(8,5))
plt.bar(sales_data["Period"], sales_data["Revenue"], color=['navy','teal','orange'])
plt.title("Celsius: Sales & Revenue Levels")
plt.ylabel("Revenue (USD millions)")
plt.xticks(rotation=0)
plt.show()

# Plot Consumer Demographics
plt.figure(figsize=(8,5))
plt.bar(demographics_data["Group"], demographics_data["Percentage (%)"], color='green')
plt.title("Celsius Consumer Demographics (% Estimates)")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=45)
plt.show()

# Plot Market Share
plt.figure(figsize=(8,5))
plt.bar(market_share_data["Brand"], market_share_data["Market Share (%)"], color='purple')
plt.title("Market Share in U.S. Energy Drink Category")
plt.ylabel("Market Share (%)")
plt.xticks(rotation=0)
plt.show()
```
