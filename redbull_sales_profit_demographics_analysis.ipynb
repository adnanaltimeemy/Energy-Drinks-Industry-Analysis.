```python
# Cell 1 — imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cell 2 — global sales / financial summary
# Values are illustrative but grounded in reported facts for 2024–2025.

global_financials = pd.DataFrame({
    "Year": [2024, 2025],
    "Cans_sold_billion": [None, 13.969],  # 2025 reported figure; 2024 value not given here.
    "Revenue_eur_billion": [11.227, 12.196],  # 2025 turnover; 2024 base for comparison.
    "Notes": [
        "Group turnover 2024, base for 2025 growth calculation",
        "Record turnover and cans sold for 2025"
    ]
})

global_financials
```

```python
# Cell 3 — UK example data, 2024 fiscal year
uk_financials = pd.DataFrame({
    "Year_end": [2024],
    "Revenue_million_gbp": [668.3],   # reported
    "Profit_million_gbp": [40.4],     # reported
    "Market_share_percent": [36.8],   # reported
    "Notes": ["UK fiscal year 2024 results"]
})

# compute a simple profit margin
uk_financials["Profit_margin_pct"] = (
    uk_financials["Profit_million_gbp"] / uk_financials["Revenue_million_gbp"] * 100
).round(2)

uk_financials
```

```python
# Cell 4 — basic consumer demographics for energy drink users (example snapshot)
demographics_snapshot = pd.DataFrame({
    "Attribute": [
        "Age under 35 share among daily users", 
        "Gen Z 18–24 share among daily users",
        "Male share among daily users",
        "Female share among daily users"
    ],
    "Percent": [
        60,   # Under 35 more likely
        64,   # Gen Z share within daily users
        34,   # Male slightly higher
        31    # Female slightly lower; remainder likely other/unspecified
    ],
    "Source": [
        "CivicScience 2025 consumer insight",
        "CivicScience 2025 consumer insight",
        "CivicScience 2025 consumer insight",
        "CivicScience 2025 consumer insight"
    ]
})

demographics_snapshot
```

```python
# Cell 5 — combine into a dict for quick reference or export
dataset = {
    "global_financials": global_financials,
    "uk_financials": uk_financials,
    "demographics_snapshot": demographics_snapshot
}

# Example: display all keys and head of each DF
for name, df in dataset.items():
    print(f"\nDataset: {name}")
    display(df.head())
```

```python
# Cell 6 — simple bar chart for global revenue 2024 vs 2025
plt.figure(figsize=(6,4))
plt.bar(
    global_financials["Year"].astype(str), 
    global_financials["Revenue_eur_billion"], 
    color=["#1e90ff", "#ff4500"]
)
plt.title("Red Bull reported global revenue (EUR billions)")
plt.ylabel("Revenue (EUR bn)")
plt.show()
```

```python
# Cell 7 — UK profit margin as text or simple chart
plt.figure(figsize=(4,4))
margin = uk_financials.loc[0, "Profit_margin_pct"]
plt.pie([margin, 100-margin], labels=[f"Profit {margin}%", "Other"], autopct='%1.1f%%', startangle=90)
plt.title("UK profit margin example, FY 2024")
plt.show()
```

```python
# Cell 8 — demographics
plt.figure(figsize=(6,4))
plt.bar(
    demographics_snapshot["Attribute"], 
    demographics_snapshot["Percent"]
)
plt.title("Snapshot: energy drink users demographics")
plt.ylabel("Percent")
plt.xticks(rotation=25, ha="right")
plt.tight_layout()
plt.show()
```

```python
global_financials.to_csv("global_financials.csv", index=False)
uk_financials.to_csv("uk_financials.csv", index=False)
demographics_snapshot.to_csv("demographics_snapshot.csv", index=False)
```
