```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Annual sales (mock data in millions USD)
sales_data = {
    "Year": [2019, 2020, 2021, 2022, 2023, 2024],
    "Sales_USA_Million_USD": [5200, 5300, 5400, 5500, 5600, 5800],
    "Sales_Global_Million_USD": [6100, 6300, 6500, 6700, 6900, 7100]
}

df_sales = pd.DataFrame(sales_data)
df_sales
```

```python
# Profit margin data (mock / estimated)
profit_data = {
    "Year": [2019, 2020, 2021, 2022, 2023, 2024],
    "Gross_Margin_%": [28, 29, 29, 30, 31, 32],
    "Net_Margin_%": [12, 12.5, 13, 13.5, 14, 14.2]
}

df_profit = pd.DataFrame(profit_data)
df_profit
```

```python
# Demographics mock-up for illustrative purposes
demographic_data = {
    "Group": ["Age 18-24", "Age 25-34", "Age 35-44", "Female", "Male"],
    "Share_%": [25, 30, 20, 35, 65]
}

df_demographics = pd.DataFrame(demographic_data)
df_demographics
```

```python
df_combined = pd.merge(df_sales, df_profit, on="Year")
df_combined
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_sales["Year"], df_sales["Sales_USA_Million_USD"], marker='o', label="USA")
plt.plot(df_sales["Year"], df_sales["Sales_Global_Million_USD"], marker='s', label="Global")
plt.title("Mountain Dew Sales Growth (20192024)")
plt.xlabel("Year")
plt.ylabel("Sales (Million USD)")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.bar(df_profit["Year"] - 0.15, df_profit["Gross_Margin_%"], width=0.3, label="Gross Margin")
plt.bar(df_profit["Year"] + 0.15, df_profit["Net_Margin_%"], width=0.3, label="Net Margin")
plt.title("Profit Margin Trends")
plt.xlabel("Year")
plt.ylabel("Margin (%)")
plt.legend()
plt.show()
```
