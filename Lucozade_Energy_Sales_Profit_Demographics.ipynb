```python
# ### Lucozade Energy: Sales, Profit Margins & Demographics
# A step-by-step dataset creation and basic analysis

# —————————————————————————————
# 1) IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# (optional) enable inline plots
%matplotlib inline

# —————————————————————————————
# 2) DEFINE DATA SOURCES (Placeholders / Example Data)

# Overall UK energy drinks market data (e.g., industry sales)
market_data = {
    'Year': [2022, 2023, 2024, 2025],
    'UK_Energy_Drink_Market_Value_USDbn': [4.8, 5.0, 5.08, 5.32],  
    # Approx from industry sources (e.g., energy drinks market estimate)
    # Actual Lucozade Energy share not publicly broken out
}

market_df = pd.DataFrame(market_data)

# Company level revenue/profit info (example placeholders)
company_data = {
    'Year': [2022, 2023, 2024],
    'Lucozade_Brand_Revenue_£m': [500, 543, 552],    # Approx revenues reported
    'Profit_After_Tax_£m': [85, 95, 105],            # Sample profits
}

company_df = pd.DataFrame(company_data)

# Consumer demographics — example survey data
demo_data = {
    'Age_Group': ['16-24', '25-34', '35-44', '45+'],
    'Male_User_%': [65, 55, 40, 25],     # energy drinks skew younger/men
    'Female_User_%': [35, 30, 25, 15],
}

demo_df = pd.DataFrame(demo_data)

# Product variant sales data (hypothetical split)
product_sales = {
    'Variant': ['Original', 'Zero/Sugar Free', 'Flavour Editions'],
    'Market_Share_%': [45, 35, 20],
}

product_df = pd.DataFrame(product_sales)

# —————————————————————————————
# 3) MERGE / PREPARE DATA
print("Market data sample")
print(market_df)

print("Company revenue/profit data")
print(company_df)

# —————————————————————————————
# 4) BASIC ANALYSIS

# Plot estimated UK energy drink market size over time
plt.figure(figsize=(8,5))
sns.lineplot(data=market_df, x='Year', y='UK_Energy_Drink_Market_Value_USDbn', marker='o')
plt.title("UK Energy Drinks Market Value (USD bn)")
plt.ylabel("Market Value (bn USD)")
plt.grid(True)
plt.show()

# Profit margin calculation for the Lucozade brand
company_df['Profit_Margin_%'] = (company_df['Profit_After_Tax_£m'] / company_df['Lucozade_Brand_Revenue_£m']) * 100
print("\nProfit margin over years:")
print(company_df[['Year','Profit_Margin_%']])

# Demographic visualization
plt.figure(figsize=(8,5))
demo_df.set_index('Age_Group')[['Male_User_%','Female_User_%']].plot(kind='bar')
plt.title("Energy Drink Use by Age & Gender (Example %)")
plt.ylabel("Percent of Respondents")
plt.xticks(rotation=45)
plt.show()

# Variant share
plt.figure(figsize=(6,4))
sns.barplot(data=product_df, x='Variant', y='Market_Share_%')
plt.title("Lucozade Energy Variant Share (Hypothetical)")
plt.ylabel("Market Share (%)")
plt.show()

# —————————————————————————————
# 5) EXPORT DATA
market_df.to_csv("energy_drinks_market.csv", index=False)
company_df.to_csv("lucozade_brand_financials.csv", index=False)
demo_df.to_csv("energy_drinks_demographics.csv", index=False)
product_df.to_csv("product_variant_data.csv", index=False)

print("Datasets exported!")
```
